## Read me

The purpose of the script is to seperate individual excel file based on the column's group
1. column variable: StudyID 
    - Individual transcript based on StudyID and condition (DIME001-CV; DIME002-DG)
2. column vairalbe: StudyID-contigency
    - Individual transcript based on StudyID, condition, and contignecy (DIME001-CV-con1; DIME002-DG-con2)
    
    
**Note:**\
definition of contigency:
- contigency [1]: the silence between two utterances from parents is less than 2s
- non-contigency [2]: the silence between two utterances from parents is more than 2s

```python
import pandas
import os
import pandas as pd
```

```python
# set constants
INPUT_DIR = "C:/Users/USER/PycharmProjects/UniqueWordCalculator/"
INPUT_FILE = "parent_speech.xlsx"
BASED_COLUMN = "StudyID_con"
OUTPUT_DIR = "C:/Users/USER/PycharmProjects/UniqueWordCalculator/text_agent_condition/parent/condition-contigency"
```

```python
df_input = pd.read_excel(INPUT_FILE)
```

```python
df_input.head()
```

```python
# subset the data
df_inputsub = df_input[["StudyID_con", "text"]]
df_inputsub.head()
```

```python
# get the column

uni_val = df_inputsub[BASED_COLUMN].unique()
uni_val
```

```python
# Filter the data basls on listed items and write to excle file in loop

for items in uni_val:
    dataframe = df_inputsub[df_inputsub[BASED_COLUMN]==items]
    path = os.path.join(OUTPUT_DIR, str(items) + ".xlsx")
    dataframe.to_excel(path, index=False)
```
